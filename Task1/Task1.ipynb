## STEP 1 : DATA CLEANING

**STEP 1.1 : REMOVE THE CANCELLED INVOICENO**

```python
cancelled_orders = sales_analysis_datas[sales_analysis_datas['InvoiceNo'].astype(str).str.startswith('C')]
cancelled_orders
```

The dataset contains 9,288 cancelled transactions, identified by invoice numbers starting with "C".

```python
# Remove cancelled invoices
sales_analysis_datas = sales_analysis_datas[
    ~sales_analysis_datas['InvoiceNo'].astype(str).str.startswith('C')
]
sales_analysis_datas
```

```python
sales_analysis_datas
```

**STEP 1.2 : REMOVE NEGATIVE QUANTITY, UNITPRICE**

```python
sales_analysis_datas = sales_analysis_datas[
    (sales_analysis_datas['Quantity'] > 0) &
    (sales_analysis_datas['UnitPrice'] > 0)
]
sales_analysis_datas
```

**STEP 1.3 : DROP MISSING CUSTOMERID**

```python
sales_analysis_datas['CustomerID'].isna().sum()
```

```python
NaN_customerID = sales_analysis_datas[sales_analysis_datas['CustomerID'].isna()]
NaN_customerID
```

The Dataset has 132220 observations with no CustomerID.

```python
sales_analysis_datas = sales_analysis_datas.dropna(subset=['CustomerID'])
sales_analysis_datas
```

**STEP 1.4 : Creating the column Revenue**

```python
sales_analysis_datas = sales_analysis_datas.copy()
```

```python
sales_analysis_datas['Revenue'] = sales_analysis_datas['Quantity'] * sales_analysis_datas['UnitPrice']
```

```python
sales_analysis_datas
```

```python
sales_analysis_datas.dtypes
```

**STEP 1.5 Convert InvoiceDate to datetime**

```python
sales_analysis_datas['InvoiceDate'] = pd.to_datetime(sales_analysis_datas['InvoiceDate'])
```

```python
sales_analysis_datas.dtypes
```

```python
sales_analysis_datas['YearMonth'] = sales_analysis_datas['InvoiceDate'].dt.to_period('M')
```

```python
sales_analysis_datas.head()
```

### üìä KPI Calculation & Aggregation

```python
# Total Revenue
total_revenue = sales_analysis_datas['Revenue'].sum()

# Total Orders
total_orders = sales_analysis_datas['InvoiceNo'].nunique()

# Total Customers
total_customers = sales_analysis_datas['CustomerID'].nunique()

# Average Order Value
avg_order_value = total_revenue / total_orders

# Top-Selling Product
top_product = sales_analysis_datas.groupby('Description')['Revenue'].sum().idxmax()

# Top Revenue Country
top_country = sales_analysis_datas.groupby('Country')['Revenue'].sum().idxmax()

# Highest Sales Month
highest_month = sales_analysis_datas.groupby('YearMonth')['Revenue'].sum().idxmax()

print("Total Revenue:", total_revenue)
print("Total Orders:", total_orders)
print("Total Customers:", total_customers)
print("Average Order Value:", avg_order_value)
print("Top Product:", top_product)
print("Top Country:", top_country)
print("Highest Sales Month:", highest_month)
```

## STEP 2 : Exploratory Business Data Analysis (EDA)

### 1. What is the overall revenue trend over time?

```python
revenue_trend = sales_analysis_datas.groupby('YearMonth')['Revenue'].sum()
revenue_trend.plot(kind='line',  marker='o',figsize=(8,4))
plt.grid()
plt.xlabel("YearMonth")
plt.ylabel("Revenue")
plt.title("Month-wise Revenue")
plt.show()
```

**Insight:**

Monthly revenue analysis shows a clear seasonal pattern. Sales remain relatively moderate during the first half of the year and begin to increase from August onwards, reaching a peak in November. This suggests strong customer demand during the year-end holiday season, followed by a noticeable decline in December.

### 2. Which products are the top-selling products?

```python
top_selling_products = sales_analysis_datas.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
top_selling_products.plot(kind='bar',figsize=(8,4),color=['blue','green','red','orange','purple',
'cyan','magenta','yellow','brown','gray'])
plt.xlabel("Product")
plt.ylabel("Quantity Sold")
plt.title("Top 10 Products by Sales Volume")
plt.show()
```

**Insight:**

PAPER CRAFT, LITTLE BIRDIE is the top-selling product by quantity, followed by MEDIUM CERAMIC TOP STORAGE JAR. These high-volume products indicate strong customer demand and are key drivers of overall sales volume.

### 3. Which product categories contribute the most to total revenue?

```python
# In this dataset, each product description can act as a product category.
category_revenue = sales_analysis_datas.groupby('Description')['Revenue'].sum().sort_values(ascending=False)

category_revenue.head(10).plot(kind='bar', figsize=(8,4))
plt.xlabel("Product (Category)")
plt.ylabel("Revenue")
plt.title("Top Product Categories by Revenue")
plt.xticks(rotation=75)
plt.show()
```

**Statement:**

Since the dataset does not contain a predefined product category field, product descriptions were treated as individual product categories to analyze their contribution to total revenue.

**Insight:**

PAPER CRAFT, LITTLE BIRDIE is the highest revenue-generating product category, followed by REGENCY CAKESTAND 3 TIER. These products are key contributors to overall revenue and represent high-value items for the business.

### 4. Which countries or regions drive the majority of sales?

```python
country_revenue = sales_analysis_datas.groupby('Country')['Revenue'].sum().sort_values(ascending = False).head(10)
country_revenue.plot(kind='bar',figsize=(8,4))
plt.xlabel("Country")
plt.ylabel("Revenue")
plt.title("Top Countries by Revenue")
plt.xticks(rotation=45)
plt.show()
```

**Insight:**

The United Kingdom contributes the majority of total revenue, significantly outperforming all other countries, indicating strong market dependence on the UK.

### Additional exploratory questions

### 5. Who are the high-value customers contributing most revenue?

```python
high_value_customer = sales_analysis_datas.groupby('CustomerID')['Revenue'].sum().sort_values(ascending = False).head(10)
high_value_customer.plot(kind='bar',figsize=(8,4))
plt.xlabel("CustomerID")
plt.ylabel("Total Revenue")
plt.title("Top 10 High-Value Customers")
plt.show()
```

**Insight:**

The top customers contribute the highest share of revenue, with CustomerID 14646 and 18102 being the leading contributors. Identifying these high-value customers helps the business focus on retention and personalized marketing strategies.

### 6. Are there seasonal or monthly sales patterns?

**Insight:**
  
Sales show a clear seasonal pattern, with revenue increasing in the second half of the year and peaking in November. This indicates strong demand during the year-end holiday season, followed by a decline in December.
